```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/Machine Learning Code Basics/ML/5_one_hot_encoding/homeprices.csv')
```

```python
df
```

```python
dummies=pd.get_dummies(df.town)
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['price'],axis='columns')
y = final.price
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,0]])
```

```python
model.score(X,y)
```
