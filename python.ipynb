**Area of the polygon**

```python
gdf_copy = gdf.copy()
gdf_copy = gdf_copy.to_crs(epsg = 32645)
gdf_copy["Areas"] = gdf_copy.geometry.area / 10**6
gdf_copy.head()
```

```python
gdf_copy.plot('Areas')
```

```python
gdf_area_filter = gdf[gdf_copy['Areas']>200]
```

```python
gdf_area_filter.plot()
```

Working with geometry(Shapely)

```python
from shapely.geometry import Polygon, LineString, Point
```

```python
p1 = Point(85,30)
p2 = Point(80,25)
l1 = LineString([p1,p2])
pol1 = Polygon([(82,25),(83,26),(82,23)])
pol1
```

```python
geoseries_shapely = gpd.GeoSeries([p1,p2,l1,pol1])
df_shapely = gpd.GeoDataFrame(geoseries_shapely)
df_shapely.columns = ['geometry']
df_shapely.crs = "EPSG:4326"
df_shapely
```

```python
df_shapely.plot()
```

```python
df_shapely.to_file(r'/content/drive/MyDrive/Collab/output/shapely.geojson', driver = 'GeoJSON')
```

Dissolve

```python
gdf_filter = gdf_copy[['DISTRICT','area','geometry']]
gdf_district = gdf_filter.dissolve(by='DISTRICT', aggfunc ='sum')
gdf_district.head()
```

```python
len(gdf_district)
```

```python
gdf_district.plot('area', cmap = 'RdYlGn_r', legend = True)
```

```python
gdf_district.to_file(r'/content/drive/MyDrive/Collab/output/districtdissolve.geojson', driver = 'GeoJSON')
```
