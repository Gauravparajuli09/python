```python
soaps = plt.imread('/content/drive/MyDrive/Python Practice/Image Classification with python/data/soaps.jpg')
```

```python
#Sobel Edge Detection
#import sobel
from skimage.filters import sobel
```

```python
#created a function to show the image with gray scale
def showimage(original,title):
    plt.imshow(original,cmap='gray')
    plt.axis('off')
    plt.title(title)
```

```python
soaps_gray = color.rgb2gray(soaps)
showimage(soaps_gray,'GrayScale')
```

```python
soap_edge = sobel(soaps_gray)
showimage(soap_edge,'Soap Edge')
```

```python
#Gaussian Filter
#import Gaussian
from skimage.filters import gaussian
gaussian_image = gaussian(soaps, multichannel = True)
showimage(soaps,"Original Image")
```

```python
showimage(gaussian_image,"Gaussian Filter")
```

```python
#resize, rotate
#import resize and rotate
from skimage.transform import rotate, rescale
```

```python
cat = plt.imread('/content/drive/MyDrive/Python Practice/Image Classification with python/data/kitty2.jpg')
cat_rotated = rotate(cat, -90)
showimage(cat_rotated, "Image Rotated")
```

```python
cat_rescaled = rescale(cat_rotated, 1/4, anti_aliasing=True, multichannel=True)
cat_rescaled1 = rescale(cat_rotated, 1/4, anti_aliasing=False, multichannel=True)
showimage(cat_rescaled, "Anit_aliasedon")
```

```python
showimage(cat_rescaled1, "Anti_aliasedoff")
```

```python
#resize proportionally
from skimage.transform import resize
```

```python
height = float(cat_rotated.shape[0]/2)
width = float(cat_rotated.shape[1]/2)
```

```python
cat_resized = resize(cat_rotated, (height, width), anti_aliasing=True)
```

```python
showimage(cat_resized,"Resized")
```

```python
def plot_comparison(original, filtered, title_filtered):
  fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, sharey=True)
  ax1.imshow(original, cmap=plt.cm.gray)
  ax1.set_title('original')
  ax1.axis('off')
  ax2.imshow(filtered, cmap=plt.cm.gray)
  ax2.set_title(title_filtered)
  ax2.axis('off')
```
