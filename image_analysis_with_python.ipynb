```python
text_page = plt.imread('/content/drive/MyDrive/Python Practice/Image Classification with python/data/text_page_original.png')
```

```python
from skimage.filters import threshold_local
```

```python
type(text_page)
text_page_gray = rgb2gray(text_page)
```

```python
soaps = plt.imread('/content/drive/MyDrive/Python Practice/Image Classification with python/data/soaps.jpg')
```

```python
#import sobel
from skimage.filters import sobel
```

```python
#created a function to show the image with gray scale
def showimage(original,title):
    plt.imshow(original,cmap='gray')
    plt.axis('off')
    plt.title(title)
```

```python
soaps_gray = color.rgb2gray(soaps)
showimage(soaps_gray,'GrayScale')
```

```python
soap_edge = sobel(soaps_gray)
showimage(soap_edge,'Soap Edge')
```
